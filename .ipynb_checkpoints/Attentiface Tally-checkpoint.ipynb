```python
import seaborn as sns
from itertools import product

df = pd.read_excel('ATTENTIFACE.xlsx')

plt.style.use('seaborn-talk')
plt.figure(figsize = (20, 15))
ax = sns.countplot(y = cols[0] , data = df)
plt.title(cols[0] , fontsize = 30)
plt.xlabel('% of respondents', fontsize = 24)
plt.ylabel('', fontsize = 24)
plt.tick_params(labelsize=18);
    
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    totals = []
    for i in ax.patches:
        totals.append(i.get_width())
    total = sum(totals)
    ax.text(width + 0.3, y + 0.38, 
            str(
                round((width/total) * 100, 2)) 
                + '%' ,
            fontsize=18)
```
